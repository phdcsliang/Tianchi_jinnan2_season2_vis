```python
import os
import numpy as np
import pycocotools.mask as maskutil
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
import cv2
```

```python
# data_path
data_root = '../data/Second_round_data/'
train_folder = data_root + 'jinnan2_round2_train_20190401/'
train_img_folder = train_folder + 'restricted/'
annFile = train_folder + 'train_restriction.json'
```

```python
# load anno
cocoGt=COCO(annFile)
imgIds=sorted(cocoGt.getImgIds())
```

```python
# get random img and annos
rand_id = np.random.choice(imgIds)
img_info = cocoGt.loadImgs(ids=int(rand_id))[0]
anno_id = cocoGt.getAnnIds(imgIds=rand_id)
annos = cocoGt.loadAnns(ids=anno_id)

# visualize
img = cv2.imread(train_img_folder + img_info['file_name'])
bg = np.zeros((img_info['height'], img_info['width'], 3)) 
plt.figure(figsize=(10, 20))
# raw img
plt.subplot(1,3,1)
plt.imshow(img)
# raw img with annos
plt.subplot(1,3,2)
plt.imshow(img)
cocoGt.showAnns(annos)

# annos
plt.subplot(1,3,3)
plt.imshow(bg)
plt.axis('off')
cocoGt.showAnns(annos)
plt.show()
```
